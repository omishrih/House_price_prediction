```python
### Model Building
```

```python
filename="drop"
names=['price','bedroom','bathroom','sqft_living','sqft_lot','floors','waterfront','view','condition','grade','sqft_above','sqft_basement','yr_built','lat','long','sqft_living15','sqft_lot15']
df1 = pd.DataFrame(drop)
# df=(drop(filename,names=names))
```

```python
X = drop[['bedrooms', 'bathrooms','sqft_living','floors','condition','lat','long']]
y = drop['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=20,random_state=1)
```

```python
model=SVC(gamma="auto")
```

```python
### Model  Fitting
```

```python
model.fit(X_train,y_train)
```

```python
result=model.score(X_test,y_test)
result
```

```python
value=[[3,2,1000,3,3,47,-122]]
predictions=model.predict(value)
print(predictions[0])
```
